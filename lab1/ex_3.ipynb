```python
import pyod.utils.data as pyoddata
import numpy as np
from sklearn.metrics import balanced_accuracy_score

contamination = 0.1

dataset = pyoddata.generate_data(
    n_train=1000, n_test=0, n_features=1, contamination=contamination
)

X_train, X_test, y_train, y_test = dataset
X = X_train.flatten()

z_score = (X - np.mean(X)) / np.std(X)

threshold = np.quantile(np.abs(z_score), 1 - contamination)

y_pred = (np.abs(z_score) > threshold).astype(int)

print("Balanced Accuracy:", balanced_accuracy_score(y_train, y_pred))
```
